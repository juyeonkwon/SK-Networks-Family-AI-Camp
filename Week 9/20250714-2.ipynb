```python
target_oid = "477"
target_aid = "0000559596"
# 쿼리 작성 ($elemMatch 사용)
# result.newsList 배열의 단일 요소가 oid와 aid 조건을 모두 만족해야 함
query = {
    "result.newsList": {
        "$elemMatch": {
            "oid": target_oid,
            "aid": target_aid
        }
    }
}

# 쿼리 실행 및 결과 확인
print(f"🔍 oid={target_oid}, aid={target_aid} 조건으로 문서를 조회합니다.")
# result_document = corpus.find_one(query)
```

```python
projection = {
    "_id": 0,  # _id 필드는 제외
    "result.newsList.$": 1
}

# 3. find_one에 query와 projection을 함께 전달합니다.
result_document = corpus.find_one(query, projection)
```
