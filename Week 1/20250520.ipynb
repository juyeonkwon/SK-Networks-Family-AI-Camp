- 사용자에게 값을 입력 받고,
- 입력값이 5보다 작으면 출력을 "걸어가세요"
- 입력값이 5~10까지 범위면 "버스"
- 입력값이 10을 초과하면 "택시"

```python
input_a = int(input("값을 입력하세요"))
if input_a < 5:
    print("걸어가세요")
elif input_a <= 10:
    print("버스")
else:
    print("택시")
```
